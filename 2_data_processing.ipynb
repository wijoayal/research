```python
#Se modifica el tipo de dato de la columna date_modified para que sea tratada como fecha
df_3["date_modified"] = pd.to_datetime(df_3["date_modified"], format="%Y-%m-%dT%H:%M:%S")
df_3['date_modified'] = pd.to_datetime(df_3['date_modified'].astype('int64'), unit='ns')
#print(df_3['date_modified'].head(5))
```

```python
#se carga el archivo con las marcas de los eventos anomalos 
#Importante deberan tener una cabecera de event y tiempo si se seguira con los siguiente
dg3=pd.read_excel('portflap_events32.xls').dropna(axis=1, how='all')
```

```python
#Se modifica el tipo de dato de la columna tiempo para que sea tratada como fecha
dg3["tiempo"] = pd.to_datetime(dg3["tiempo"], format="%Y-%m-%dT%H:%M:%S")
dg3['tiempo'] = pd.to_datetime(dg3['tiempo'].astype('int64'), unit='ns')
```

```python
#Se itera en el rango del dataset de portflap y tambien por el de eventos con el fin de si una marca de tiempo del dataset de 
#port flap corresponda a una de las marcas se la catalogue como anomalia al colocar un numero 1 
#Se realizo de esta manera debido a como estaba estructurado el documento que recopilaba las marcas de tiempo de eventos anomalos
d_df3=df_3
d_df_gt3=dg3

for i in range(len(d_df3)):
    d_df3.loc[i,"Anomaly"] = "0";
#sera 1
for i in range(len(d_df3)):
    for j in range(0,len(d_df_gt3),2):
        if((d_df3.loc[i,"date_modified"]>d_df_gt3.loc[j,"tiempo"]) & (d_df3.loc[i,"date_modified"]<d_df_gt3.loc[j+1,"tiempo"])):
            d_df3.loc[i,"Anomaly"] = "1";
```
